```python
def scheme(i):
    if i in ["Ministry of Labour and Employment","Ministry of Labour & Employment"]:
        return "Labour"
    elif i in ["E.P.F.O","Bank Saving Account","P.F","Vidhwa Pension Yojana","Bank Account open","Saving Account","Saving Acconunt"]:
        return "Bank/Pension services"
    elif i=="Matru Vandana Yojna":
        return "Mother&Child"
    elif i in ["School Admision","Scholarship Form","Collage Scholarship"]:
        return "Educational"
    elif i in ["Kanya Vivha Sahayta Yojna","Sukanya Yojana"]:
        return "Girl child"
    elif i=="Kishan Samman Nidhi Yojna":
        return "Agriculture"
    elif i=="Ministry of Health and Family Welfare":
        return "Health"
    elif i=="SHG Group Loan":
        return "SHG loan"
    else:
        return i
    
    
```

```python
df["Scheme categories"]=df["Scheme"].apply(scheme)
```

```python
df["ENTITLEMENT / CLAIMS NAME"].value_counts()
def entitlement_category(i):
    if i in ["Aayushman Card","Aayushman Card Download","Health Card","Aayushman Card download"]:
        return "Health"
    elif i=="E-Shram Card":
        return "Labour"
    elif i in ["Kishan Samman Nidhi E-KYC","PM kisan kyc","Khaadhany Parchi Download"]:
        return "Agriculture"
    elif i in ["P.F.Claim","P.F Claim","Oldage Pansion","Bank account open","Saving join Account Open","New Pan Card",
               "Pan Card New","New Pan card","Vidhva Pansion","Vidvha pension","Samagra ID"]:
        return "Pension/Bank/PAN/Social security"
    elif i in ["Aadhar Card","Aadhar Card New","Aadhar card download"]:
        return "Aadhar"
    elif i in ["Vivah sahaayta yojana","Suknya samridhi yojna","Janani Suraksha Yojana"]:
        return "Women/Girl child"
    elif i in ["Cast Cartificate","Income certificate","Caste certificate","Domicile Certificate",
              "Domicile cetificate","Income Certificate"]:
        return "Certificates"
    elif i in ["Aawas Yojna"]:
        return "Housing"
    elif i in ["School Admission","Collage Scholarship","Shramodaya Vidyalaya form fill up"]:
        return "Education"
    elif i=="Sochalay Yojana":
        return "Sanitation/Hygiene"
    else:
        return i
    
```

```python
df.columns
df["ENTITLEMENT / CLAIMS NAME"]=df["ENTITLEMENT / CLAIMS NAME"].apply(entitlement_category)
```

# MACHINE LEARNING

```python
for i in df.columns:
    if df[i].isnull().sum()*100/df.shape[0]>85:
        df.drop(i,axis=1,inplace=True)
```

```python
df.drop(["Date","NAME OF BENEFICIARIES"],axis=1,inplace=True)
```

```python
df.drop("Remark",axis=1,inplace=True)
```

```python
df.nunique()
def sex(i):
    if i=="M":
        return 1
    else:
        return 0
```

```python
df["M/F"]=df["M/F"].apply(sex)
```

```python
df["VILLAGE NAME"].value_counts()
def vil(i):
    if i=="Malanpur Centre":
        return 2 
    elif i in ["Tilori","Lahachura ka pura","Kushwah colony","Singhwari","Samta nagar","Harirampura"]:
        return 1
    else:
        return 0
```

```python
df["VILLAGE NAME"]=df["VILLAGE NAME"].apply(vil)
```

```python
df["Type of outreach"]=df["Type of outreach"].fillna(df["Type of outreach"].mode()[0])
```

```python
df["Type of outreach"].value_counts()
def out(i):
    if i=="Service Camp":
        return 1
    else:
        return 0
```

```python
df["Type of outreach"]=df["Type of outreach"].apply(out)
```

```python
df["Scheme categories"]=df["Scheme categories"].fillna(df["Scheme categories"].mode()[0])
```

```python
df["Scheme categories"].value_counts()
def sch_cat(i):
    if i in ["Agriculture","Bank/Pension services","Health"]:
        return 1
    else:
        return 0
```

```python
df["Scheme categories"]=df["Scheme categories"].apply(sch_cat)
```

```python
df.drop("Scheme",axis=1,inplace=True)
```

```python
df["ENTITLEMENT / CLAIMS NAME"].value_counts()
df["ENTITLEMENT / CLAIMS NAME"]=df["ENTITLEMENT / CLAIMS NAME"].fillna(df["ENTITLEMENT / CLAIMS NAME"].mode()[0])
```

```python
df["ENTITLEMENT / CLAIMS NAME"].value_counts()

def ent(i):
    if i=="Health":
        return 2
    if i in ["Labour","Pension/Bank/PAN/Social security","Agriculture","Aadhar"]:
        return 1
    else:
        return 0
```

```python
df["ENTITLEMENT / CLAIMS NAME"]=df["ENTITLEMENT / CLAIMS NAME"].apply(ent)
```

```python
df.head() 
df["Process completed"]
df["Process completed"]=df["Process completed"].fillna(df["Process completed"].mode()[0])
```

```python
df["Process completed"].value_counts()
def proc(i):
    if i=="Done":
        return 1
    else:
        return 0
    
```

```python
df["Process completed"]=df["Process completed"].apply(proc)
```

```python
df["Religion"].value_counts()

def re(i):
    if i=="Others":
        return 1
    else:
        return 0
```

```python
df["Religion"]=df["Religion"].apply(re)
```

```python
df["COMMUNITY - SC/ST/OBC & General"]=df["COMMUNITY - SC/ST/OBC & General"].fillna(df["COMMUNITY - SC/ST/OBC & General"].mode()[0])
```

```python
df.isnull().sum()
```

```python
pd.get_dummies(df[["COMMUNITY - SC/ST/OBC & General","Month"]],drop_first=True)
df_final=pd.concat([df.drop(["COMMUNITY - SC/ST/OBC & General","Month"],axis=1),pd.get_dummies(df[["COMMUNITY - SC/ST/OBC & General","Month"]],drop_first=True)],axis=1)
```

```python
df_final.head()
X=df_final.drop("Process completed",axis=1)
y=df_final["Process completed"]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
cross_val_score(log,X,y,cv=5,scoring="f1")
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.25,random_state=10)
model=log.fit(xtrain,ytrain)
pred=model.predict(xtest)
print(classification_report(ytest,pred))
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
model=xgb.fit(xtrain,ytrain)
pred=model.predict(xtest)
print(classification_report(ytest,pred))
print(cross_val_score(xgb,X,y,cv=5,scoring="f1"))

```
